```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# List of items
items = ['Item 1', 'Item 2', 'Item 3']

# Action to perform when button is clicked
def on_button_click(button, item):
    clear_output(wait=True)
    display_ui(items)
    print(f"Button for {item} clicked!")

# Create a simple UI to display items with buttons
def display_ui(items):
    for item in items:
        button = widgets.Button(description=f"Click {item}")
        button.on_click(lambda b, item=item: on_button_click(b, item))
        display(widgets.HBox([widgets.Label(item), button]))

display_ui(items)
```

```python
import re
from collections import Counter

def get_sorted_words_by_frequency(strings):
    # Use a regular expression to split strings into words using non-alpha characters as delimiters
    words = [word.lower() for s in strings for word in re.split(r'[^a-zA-Z]+', s) if word]
    
    # Count the frequency of each word
    word_counts = Counter(words)
    
    # Sort words by frequency in decreasing order
    sorted_words = sorted(word_counts.keys(), key=lambda x: (-word_counts[x], x))
    
    return sorted_words

# Test
strings = ["Hello, world!", "How are you, world?", "Hello, user!", "This is fun", "what's up doc", "you are smelly"]
print(get_sorted_words_by_frequency(strings))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_series(arrays, names):
    """
    Plot a series of 1D numpy arrays with different colors and labels.

    Parameters:
    - arrays: List of 1D numpy arrays.
    - names: List of names corresponding to each numpy array.
    """
    # Check if the lengths of arrays and names match
    if len(arrays) != len(names):
        raise ValueError("The number of arrays and names should match.")
    
    # Plot each array
    for array, name in zip(arrays, names):
        plt.plot(array, label=name)
    
    # Display legend, grid and show the plot
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
arrays = [np.random.rand(10), np.random.rand(10), np.random.rand(10)]
names = ["Series 1", "Series 2", "Series 3"]
plot_series(arrays, names)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_series(arrays, names, bar_chart=False):
    """
    Plot a series of 1D numpy arrays with different colors and labels.

    Parameters:
    - arrays: List of 1D numpy arrays.
    - names: List of names corresponding to each numpy array.
    - bar_chart: Boolean indicating whether to plot as bar chart. Default is False (line plot).
    """
    if len(arrays) != len(names):
        raise ValueError("The number of arrays and names should match.")

    if bar_chart:
        # Bar chart plotting
        bar_width = 0.8 / len(arrays)  # Calculate width for each bar series
        for idx, (array, name) in enumerate(zip(arrays, names)):
            # Calculate the positions of bars for each series
            positions = np.arange(len(array)) + idx * bar_width
            plt.bar(positions, array, width=bar_width, label=name)
        plt.xticks(np.arange(len(arrays[0])) + 0.4, np.arange(len(arrays[0])))
    else:
        # Line plot
        for array, name in zip(arrays, names):
            plt.plot(array, label=name)

    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
arrays = [np.random.rand(10), np.random.rand(10), np.random.rand(10)]
names = ["Series 1", "Series 2", "Series 3"]
plot_series(arrays, names, bar_chart=False)
```
